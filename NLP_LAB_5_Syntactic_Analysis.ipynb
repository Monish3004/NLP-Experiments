```python
#Check Grammar of sentence ,
#POS
#How one word is linked to another

import spacy

# Load spaCy's English language model
nlp = spacy.load('en_core_web_sm')

# Sample sentence
sentence = input("Enter a sentence to visualize its structure: ")

# Process the sentence using spaCy
doc = nlp(sentence)

# Print POS tags and dependency relations
print(f"{'Word':<12}{'POS':<10}{'Dependency Relation':<20}{'Head':<12}")
print("-" * 50)
for token in doc:
    print(f"{token.text:<12}{token.pos_:<10}{token.dep_:<20}{token.head.text:<12}")
```

```python
from spacy import displacy

displacy.render(doc, style='dep', jupyter=True)
```

```python
import nltk
import matplotlib.pyplot as plt
from nltk import CFG
from nltk.tree import Tree

nltk.download('punkt')

grammar = CFG.fromstring("""
    S -> NP VP
    NP -> Det N | Det N PP
    VP -> V NP | V NP PP
    PP -> P NP
    Det -> 'the' | 'a'
    N -> 'cat' | 'dog' | 'mat' | 'rat'
    V -> 'chased' | 'sat'
    P -> 'on' | 'with'
""")

# Define a sentence to parse
sentence = "the cat chased the rat".split()

# Create the parser
parser = nltk.ChartParser(grammar)

# Generate and plot the parse trees
for tree in parser.parse(sentence):
    print(tree)
    tree.pretty_print()  # Text-based print
```

```python
import nltk
from nltk import CFG

# Define the grammar for parsing
grammar = CFG.fromstring("""
  S -> NP VP
  NP -> Det N | Det N PP | 'I'
  VP -> V NP | V NP PP
  PP -> P NP
  Det -> 'a' | 'the'
  N -> 'dog' | 'cat' | 'park'
  V -> 'saw' | 'chased'
  P -> 'in' | 'with'
""")

# Create the parser
parser = nltk.ChartParser(grammar)

# I saw a dog in the park
sentence = input("Enter a sentence: ").split()

# Generate the parse tree
try:
    for tree in parser.parse(sentence):
        #print(tree)
        tree.pretty_print()
except ValueError:
    print("The sentence does not conform to the given grammar.")
```
